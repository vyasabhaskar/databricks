```python
# create a dictionary and get the dictionary sorted based on values
import operator 
dict1 = {1:20, 2:10, 3:30, 4:15, 5: 14}

print("original dictionary: ", dict1)

print("sorting starts")

sorted_Dict = sorted(dict1.items(), key = operator.itemgetter(1))

print("Sorted Order in Ascending Order: ", sorted_Dict)

sorted_Dict_reverse = sorted(dict1.items(), key = operator.itemgetter(1), reverse="true")

print("Sorted Order in Descending Order: ", sorted_Dict_reverse)
```

```python
#Generate Dictionary of Numbers and Their Squares

def dict_squares(n):
    return n*n

dict_squares = {x:dict_squares(x) for x in range(1,11)}
print(dict_squares)
```
