```python
file_path = '/Volumes/databricks_airbnb_sample_data/v01/sf-listings/sf-airbnb.csv'
df = pd.read_csv(file_path)
```

```python
df.head(3)
```

```python
df.tail(3)
```

```python
# rename column
df = df.rename(columns={"neighbourhood":"neighborhood"})
df.head(3)
```

```python
# using filters 
filtered_df = df[df["host_is_superhost"] == 't']
filtered_df.head(3)
```

```python
# pandas boolean operator
filtered_df = df[(df["host_is_superhost"] == 't') & (df["number_of_reviews"]> 150)]
filtered_df.head(3)
```

```python
# aggregate functions

print(df["number_of_reviews"].max())
print(df["number_of_reviews"].min())
print(df["number_of_reviews"].mean())
```

```python
# describe ()
df[["number_of_reviews","bedrooms"]].describe().round(2)
```

```python
# group by
groupedby_df = df.groupby(["neighborhood"])[["bedrooms","bathrooms"]].mean().head(10).round(2)
print(groupedby_df)
```

```python
groupedby_df.columns
groupedby_df.reset_index()
```

```python
# sorting
groupedby_df.sort_values(["bedrooms"]).reset_index()
```
