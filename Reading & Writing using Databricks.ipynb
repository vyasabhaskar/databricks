```python
# read the csv file
df = spark.read.format("csv").option("sep","|").option("header","true").option("inferschema","true").load("/Volumes/databricks_simulated_e_commerce_clickstream_data/v01/raw/sales-csv/")
display(df)
```

```python
df.printSchema()
```

```python
df.schema
```

```python
write_file_path = "/Volumes/workspace/default/csvfiles/sales_parquet"
df.write.format("parquet").mode("overwrite").save(write_file_path)
```

```python
# show the files in the directory
display(dbutils.fs.ls(write_file_path))
```

```python
# write dataframe data to a table
df.write.saveAsTable("sales_df_table")
```

```python
# use writeTo method which invokes DataFramewriterV2 
df.writeTo("workspace.default.sales_df_writeto_table").createOrReplace()
```
