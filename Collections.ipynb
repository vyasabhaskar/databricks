count of words in a string 

```python
sentence = "Hello World Hello World Hello"

# create empty dictionary 
counts = {}

#split the sentence into words
words = sentence.split()
print(words)

for word in words:
    if word in counts:
        counts[word]+=1
    else:
        counts[word] = 1
    print(counts)
```
