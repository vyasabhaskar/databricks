```python
# Maximum and Minimum K elements in Tuple
import heapq
t = (2, 4, 3, 6, 8, 11, 1, 9)
k=2

s = heapq.nsmallest(k, t)
l = heapq.nlargest(k, t)

print(s)
print(l)
```

```python
# Removing duplicates from tuple 
# using tuple() + set()

# initialize tuple
test_tup = (1, 3, 5, 2, 3, 5, 1, 1, 3)

dup = set()
unique_Tup = ()

for i in  test_tup:
    if i not in dup:
        dup.add(i)
        unique_Tup += (i,)
print(unique_Tup)

# another method is to use set() and tuple()

print(tuple(set(test_tup)))
```
