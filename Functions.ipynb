```python
def func_dollar_to_rupee_conversion():
    print(10 * 88)

func_dollar_to_rupee_conversion()   
```

```python
# with conversion rate as parameter to the function
def func_dollar_to_rupee_conversion(conversion_rate):
    print(10 * conversion_rate)

func_dollar_to_rupee_conversion(88)  
```

```python
# with both conversion rate and dollar amount as parameter to the function
def func_dollar_to_rupee_conversion(dollar_amount,conversion_rate):
    print(dollar_amount * conversion_rate)

func_dollar_to_rupee_conversion(10,88)  
```

```python
# with both conversion rate and dollar amount as parameter to the function
def func_dollar_to_rupee_conversion(dollar_amount:float, conversion_rate:float):
    print(dollar_amount * conversion_rate)

func_dollar_to_rupee_conversion(10.0,88.5)  
```

```python
# with both conversion rate and dollar amount as parameter to the function and default conversion rate. 
def func_dollar_to_rupee_conversion(dollar_amount:float, conversion_rate:float = 88.5):
    print(dollar_amount * conversion_rate)

func_dollar_to_rupee_conversion(10.0)  
```

```python
# with default value and assigning output to a variable 
def func_dollar_to_rupee_conversion(dollar_amount:float, conversion_rate:float = 88.5):
    return dollar_amount * conversion_rate

rupee_amount = func_dollar_to_rupee_conversion(10.0)
print(rupee_amount)
```

```python
# with default value and assigning output to a variable 
def func_dollar_to_rupee_conversion(dollar_amount:float, conversion_rate:float = 88.5):
    return dollar_amount * conversion_rate

rupee_amount = func_dollar_to_rupee_conversion(10.0,84.8)
print(rupee_amount)
```
