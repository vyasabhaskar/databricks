```python
# __init__ method
class MyClass:
    def __init__(self, name, age):
        self.name = name
        self.age = age

#create Myclass object 
p1 = MyClass("Bhaskar","37")

print(p1.name)
p1.age
```

```python
# __init__ method
class MyClass:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def greet(self):
        print ("Hello My Name is: " + self.name)

#create Myclass object 
p1 = MyClass("Bhaskar","37")
p2 = MyClass("Mani","37")

p1.greet()
p2.greet()
```
